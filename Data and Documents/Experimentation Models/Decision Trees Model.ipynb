##Handling the missing values

```python
df.isnull().sum()
```

```python
df['Price'].fillna(df['Price'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

##Converting the Date format to the int

```python
import datetime
```

```python
df['Date']=pd.to_datetime(df.Date,format='%Y-%m-%d')
df.dtypes
```

```python
df['year']=df['Date'].dt.year
df['month']=df['Date'].dt.month
df['day']=df['Date'].dt.day
```

```python
df['date']=(df['year']*100+df['month'])*100+df['day']
```

```python
df['date'].head()
```

## Splitting the model to train and test

```python
from sklearn.model_selection import train_test_split
```

```python
X=df['date']
y=df['Price']
```

```python
X.head()
```

```python
y.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=24)
```

## Decision Tree Model

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(np.array(X_train).reshape(-1,1),y_train)
```

## Predicting the values

```python
pred=regressor.predict(np.array(X_test).reshape(-1,1))
```

```python
pred
```

```python
y_test = np.array(y_test)
```

```python
y_test
```

## Mean_Squared_Error and r2_score

```python
from sklearn.metrics import mean_squared_error,r2_score
import math
print(math.sqrt(mean_squared_error(y_test,pred)))
print(r2_score(y_test,pred))
```

```python
print(mean_squared_error(y_test,pred))
```
